```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/DataAnalyst_Assesment_Dataset1.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
#check Messing values in dataset
print(df.isnull().sum())
```

```python
df['Booking Date']=pd.to_datetime(df['Booking Date'])
df['Time Slot']=pd.to_datetime(df['Time Slot'],format='%H:%M')
```

```python
df.info()
```

```python
df.head()
```

```python
df.fillna({'Time Slot': 0}, inplace=True)
```

```python
df.head()
```

```python
df = df.drop('Subscription Type', axis=1)
```

```python
df.info()
```

```python
df.fillna({'Duration (mins)': 0}, inplace=True)
df.dropna(subset=['Customer Email','Customer Phone'],inplace=True)
print(df.isnull().sum())
```

```python
df.head()
```

```python
df.fillna({'Class Type': 'Unknown','Instructor':'Unknown','Facility':'Unknown','Theme':'Unknown'}, inplace=True)
print(df.isnull().sum())
```

```python
df.head()
```

```python
df["Revenue"] = df["Price"] * df["Duration (mins)"]
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df = df.copy()
df['Month'] = df['Booking Date'].dt.month_name()
df['Monthly Revenue'] = df.groupby('Month')['Revenue'].transform('sum')
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df=df[df["Duration (mins)"] > 0]
```

```python
df.info()
```

```python
df.describe()
```

```python

# Count bookings per service type
service_counts = df["Service Type"].value_counts()

plt.figure(figsize=(5, 3))
sns.barplot(x=service_counts.index, y=service_counts.values, hue=service_counts.index, palette="viridis", legend=False)

plt.xticks(rotation=45)
plt.xlabel("Service Type")
plt.ylabel("Number of Bookings")
plt.title("Most Popular Services")
plt.show()
```

```python
# Group revenue by month
monthly_revenue = df.groupby("Month")["Revenue"].sum()

plt.figure(figsize=(5, 3))
sns.lineplot(x=monthly_revenue.index, y=monthly_revenue.values, marker="o", color="b")
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.title("Revenue Trend Over Time")
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.histplot(df["Revenue"], bins=30, kde=True, color="blue")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.title("Revenue Distribution")
plt.show()
```

```python
df.to_csv("Cleaned_Data.csv", index=False)
print("Cleaned dataset saved successfully!")
```
